### 十分統計量とは

$X_1,...,X_n$をn個の独立な成功確率$p$のベルヌーイ変数とする。<br>
このとき、$ Y= \sum^n_{i=1} X_i$とすると、$Y$は二項分布$Bin(n,p)$に従う。<br>

ここで$p$を推定するとして、統計量$Y$について考える。<br>

$X=(X_1,...,X_n)$の代わりに$Y$のみを知った時に$X$を知った時と比較して失われてた情報は$Y$回の成功と$n-Y$回の失敗がどのように並んでいたかである。<br>
ここで$p$を知ることに関して成功と失敗の並び順がどのように関わるかを考える。<br>
このことは次のようにYを与えた時のXの条件付き確率を考えればわかる。<br>

$Y=y$が与えられたという条件のもとで特定の$x=(x_1,x_2,...,x_n)$が観測される条件付き確率は、<br>
$y=x_1+x_2+ \cdots +x_n$に注意して<br>

$$
\begin{eqnarray}
P(X_1=x_1,...,X_n=x_n|Y=y) & = & \frac{P(X_1=x_1,...,X_n=x_n,Y=y)}{P(Y=y)}\\
& = & \frac{P(X_1=x_1,...,X_n=x_n)}{P(Y=y)} \\
& = & \frac{p^{\sum^n_{i=1}x_i}\:(1-p)^{n-\sum^n_{i=1}x_i}}{\binom{n}{y}p^y(1-p)^{n-y}} \\
& = & 1/\binom{n}{y}
\end{eqnarray}
$$

となる。<br>
このとこより、$p$の値に関わらず、$\binom{n}{y}$通りの成功と失敗の並び方は全て同様に確からしいことがわかる。<br>
よって$Y=y$を考えればよく、成功と失敗の並び自体は$p$に関する情報を含んでいないので、$p$を推定するのに必要ないことがわかる。<br>

もともとの標本$X=(X_1,X_2, ..., X_m)$を知らなくとも、<br>
統計量$T=T(X_1, X_2, ..., X_n)$を知れば未知パラメータ$\theta$に関する推測については情報を何も失わない(これを知れば$\theta$に関する推測については十分)<br>
という意味で、十分統計量を次のように定義する<br>

---
定義(十分統計量)
$$
k個の統計量T=(T_1, T_2, ..., T_k)がパラメータ\thetaに関する十分統計量であるとは、\\
T(X)=tを与えた時のX=(X_1,X_2,...,X_n)の条件付き分布が\thetaに依存しないことである。\\
すなわち、\\
P(X=x|T(X)=t,\theta) = P(X=x|T(X)=t)を満たすことである。
$$

$$
\Big(\thetaの値がわかっているとき、T(X)=t \; の下でのX=xとなる条件付き確率 \quadと \quad \theta\:の値が未知のときの条件付き確率が等しい\Big)
$$

---

### 分解定理

上の定義によると、$T$が十分統計量であることを示すには、$T=t$を与えた時の$X$の条件付き分布を調べる必要がある。<br>
しかし、次の十分統計量に関する分解定理を用いれば、$T$が十分統計量であることが言える。<br>

---
定理(分解定理)

$$
Xを離散または連続の確率変数とする。P_{\theta}をXの確率質量関数または確率密度関数とすると、 \\
T(X)=(T_1(X), T_2(X),...,T_k(X))\;が\;\theta\;に関する十分統計量であるための必要十分条件は \\
P_\theta が\; P_\theta = g_\theta(T(X))h(x)\;\:の形に分解できることである。
$$

$$
\Big(つまり、P_\thetaを変形して、T(X)を含んだ形の関数\; g_\theta(T(X)) \; と\theta を含まない関数\; h(x) \;の積にできれば、Tが十分統計量であることが言える \Big)
$$
