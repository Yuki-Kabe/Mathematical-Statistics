### 十分統計量とは

$X_1,...,X_n$をn個の独立な成功確率$p$のベルヌーイ変数とする。<br>
このとき、$ Y= \sum^n_{i=1} X_i$とすると、$Y$は二項分布$Bin(n,p)$に従う。<br>

ここで$p$を推定するとして、統計量$Y$について考える。<br>

$X=(X_1,...,X_n)$の代わりに$Y$のみを知った時に$X$を知った時と比較して失われてた情報は$Y$回の成功と$n-Y$回の失敗がどのように並んでいたかである。<br>
ここで$p$を知ることに関して成功と失敗の並び順がどのように関わるかを考える。<br>
このことは次のようにYを与えた時のXの条件付き確率を考えればわかる。<br>

$Y=y$が与えられたという条件のもとで特定の$x=(x_1,x_2,...,x_n)$が観測される条件付き確率は、<br>
$y=x_1+x_2+ \cdots +x_n$に注意して<br>

$$
\begin{eqnarray}
P(X_1=x_1,...,X_n=x_n|Y=y) & = & \frac{P(X_1=x_1,...,X_n=x_n,Y=y)}{P(Y=y)}\\
& = & \frac{P(X_1=x_1,...,X_n=x_n)}{P(Y=y)} \\
& = & \frac{p^{\sum^n_{i=1}x_i}\:(1-p)^{n-\sum^n_{i=1}x_i}}{\binom{n}{y}p^y(1-p)^{n-y}} \\
& = & 1/\binom{n}{y}
\end{eqnarray}
$$

となる。<br>
このとこより、$p$の値に関わらず、$\binom{n}{y}$通りの成功と失敗の並び方は全て同様に確からしいことがわかる。<br>
よって$Y=y$を考えればよく、成功と失敗の並び自体は$p$に関する情報を含んでいないので、$p$を推定するのに必要ないことがわかる。<br>

もともとの標本$X=(X_1,X_2, ..., X_m)$を知らなくとも、<br>
統計量$T=T(X_1, X_2, ..., X_n)$を知れば未知パラメータ$\theta$に関する推測については情報を何も失わない(これを知れば$\theta$に関する推測については十分)<br>
という意味で、十分統計量を次のように定義する<br>

---
定義(十分統計量)
$$
k個の統計量T=(T_1, T_2, ..., T_k)がパラメータ\thetaに関する十分統計量であるとは、\\
T(X)=tを与えた時のX=(X_1,X_2,...,X_n)の条件付き分布が\thetaに依存しないことである。\\
すなわち、\\
P(X=x|T(X)=t,\theta) = P(X=x|T(X)=t)を満たすことである。
$$

$$
\Big(\thetaの値がわかっているとき、T(X)=t \; の下でのX=xとなる条件付き確率 \quadと \quad \theta\:の値が未知のときの条件付き確率が等しい\Big)
$$

---

### 分解定理

上の定義によると、$T$が十分統計量であることを示すには、$T=t$を与えた時の$X$の条件付き分布を調べる必要がある。<br>
しかし、次の十分統計量に関する分解定理を用いれば、$T$が十分統計量であることが言える。<br>

---
定理(分解定理)

$$
Xを離散または連続の確率変数とする。P_{\theta}をXの確率質量関数または確率密度関数とすると、 \\
T(X)=(T_1(X), T_2(X),...,T_k(X))\;が\;\theta\;に関する十分統計量であるための必要十分条件は \\
P_\theta が\; P_\theta = g_\theta(T(X))h(x)\;\:の形に分解できることである。
$$

$$
\Big(つまり、P_\thetaを変形して、T(X)を含んだ形の関数\; g_\theta(T(X)) \; と\theta を含まない関数\; h(x) \;の積にできれば、Tが十分統計量であることが言える \Big)
$$

---

### 統計的決定理論の枠組みにおける十分統計量、ラオブラックウェルの定理

次に統計的決定理論の枠組みの中で十分統計量の意味、およびラオブラックウェルの定理を論じる。<br>
再掲するとパラメータ$\theta$に関する十分統計量とは元々の標本$X=(X_1,...,X_n)$を知らなくとも、統計量$T=T(X_1,...,X_n)$を知れば未知パラメータ$\theta$に関する推測については情報を何も失わない統計量のことであった。<br>
統計的決定理論では統計家の目的は未知パラメータ$\theta$に関する決定$d$を標本$X$から行う($\delta(X)$の値を決定する)ことであり、先ほどまでの議論を踏まえれば決定関数$\delta$は十分統計量のみに依存するもの( = $\delta(T)$ )を考えればよいことがわかる。($X_i$などに着目する必要はないということ)

ここである推定量$\delta(X)$よりも十分統計量$T$を与えたときの$\delta(X)$の条件付き期待値、$\delta^*(T)=E_\theta[\delta(X)|T]$の方がよりよいことを示すラオブラックウェルの定理を紹介する。<br>

---
定理(ラオブラックウェルの定理)

$$
\delta^*(T)=E_\theta[\delta(X)|T]とする
$$

$$
このとき、任意の\thetaに対して、R(\theta, \delta^*) \leq R(\theta, \delta)が成立する.\\
\Big( 等号成立はP_\theta(\delta(X)=\delta^*(T)) = 1となるとき \Big)
$$

---

ラオブラックウェルの定理はどんな推定量$\delta(X)$に対しても、それがTの関数でなければTを与えた時の条件付き期待値$E[\delta(X)|T]$により、<br>
改良されることを示している。推定量$\delta$がTの関数ならば、条件付き期待値をとっても変わらない(等号の部分)

この定理の仮定として、損失関数$L(\theta, d)$が$d$に関して、凸であり、$R(\theta, \delta) < \infty$であることが挙げられているが、<br>
5章の例で挙げた$L(\theta, d)=(\theta-d)^2$、$L(\theta, d)=|\theta-d|$はこれを満たす(中級教科書レベルならこの仮定はあまり気にしなくて良いはず)



### 完備十分統計量

十分統計量が完備性とよばれる性質を持つ場合には、十分統計量に基づく推測についてより強い結論を導くことができる。<br>
推定論における一様最小分散不偏推定量($UMVUE$)の議論などがその例である。<br>
完備十分統計量は指数分布族に対する十分統計量の形で現れることが多い。

---
定義(完備十分統計量)

$$
十分統計量T(X)が完備であるとは、任意の\thetaに対して、E_\theta[g(T)]=0である関数g(\cdot)は定数0に限ることである。
$$

$$
つまり、^\forall \theta,\:\: E_\theta[g(T)]=0 \Rightarrow g(T) \equiv 0 が成り立つならば、Tは完備である。
$$

### 最小十分統計量

完備十分統計量と関連のある概念として、最小十分統計量の概念がある。<br>
標本の情報からこれ以上情報を縮約することのできない十分統計量を最小十分統計量という。<br>

例としてコイン投げを取り上げる。試行総数を2nとし、前半のn回の表の数を$Y_1$、後半のn回の表の数を$Y_2$と表す。<br>
このとき、組$(Y_1,Y_2)$は十分統計量である。<br>
また表の総数$Y=Y_1+Y_2$とすれば$Y$も十分統計量である。この場合、$(Y_1,Y_2)$は十分統計量であるが、さらに情報を縮約して$Y$としても十分性は失われない。従って十分性という観点から見れば$(Y_1,Y_2)$にはまだ情報があると考えることができる。

最小十分統計量の概念は技術的であり、この本では今後必要とならないので概念を抑えておく程度でよい
最小十分統計量の数学的定義は以下のようになる。(弱い意味での最小十分、強い意味での最小十分)

---
定義(弱い意味での最小十分)
$$U=g(T)が十分統計量\Rightarrow gが単射(1対1)のとき、\\
Tを弱い意味での最小十分統計量という$$

---


この定義は十分性を失うことなくTの情報をさらに縮約することは不可能であることを表している<br>
( TからUへの写像が1対1であるからTの情報を縮約されていない )

---
定義(強い意味での最小十分)
$$任意の十分統計量Sに対してあるhが存在して、T=h(S)となるとき、\\
Tを強い意味での最小十分統計量という$$

---

この定義は任意の十分統計量の情報をTに縮約できることを表している。
