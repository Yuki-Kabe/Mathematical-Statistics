<font size=4>標本空間とは</font><br>
4章で論じたように統計的推測は標本$X=(X_1,X_2,...,X_n)$に基づいて行われる.<br>
Xの実現値の属する集合を標本空間といい、$\mathcal{X}$と表す<br>
(ex) n次元の確率変数$X=(X_1,X_2,...,X_n)$の場合は、標本空間$\mathcal{X}$はn次元ユークリッド空間 $\mathbb{R}^n$と取れば良い

<font size=4>母数、母数空間とは</font><br>
統計的推測においては$X \in \mathcal{X}$の従う分布族、つまり統計的モデルを仮定する。<br>
統計的推測において考えている分布族の分布を$P_\theta$で表す。ここで母数$\theta$は分布、分布族のパラメータである。<br>
$\theta$の取りうる値の集合を$\Theta$で表し、これを母数空間と呼ぶ。<br>

(ex) 分布族を正規分布とすればは2次元データ$\theta=(\mu, \sigma^2)$であり、母数空間は$\Theta = \{ (\mu,\sigma^2)|-\infty<\mu<\infty,\:0<\sigma^2<\infty\}$となる

推定を行う母数の値は、未知数である。その意味でこれを未知母数と呼ぶ。ただし、標本Xの分布は$\theta \in \Theta$のいずれかの値の分布$P_{\theta_0}$に一致していると考える。$\theta_0$をの真の値という。<br>
以上の記法を用いて分布族は$\{ P_\theta|\theta \in \Theta \}$と表される。

母数空間はどの母数を未知と考えているかに依存することに注意する。<br>
例えば正規分布の場合、簡単のために母分散の値が既知であると仮定して統計的推定を行う場合がある。<br>
このとき考えている分布族の母数空間は$\Theta = \{\mu|-\infty<\mu<\infty\}$である。<br>
この場合、$\mu$を未知母数、$\sigma^2$を既知母数と呼んで区別する。

<font size=4>決定、決定空間について</font><br>
統計的な手法を用いて未知母数について何らかの判断を行うものを統計家と呼ぶことにする。<br>
得られたデータに基づいて統計家のおこなう判断あるいは行動を決定$d$と呼ぶことにする。<br>
統計家の取りうる決定dの集合を決定空間とよび$D$と表す。<br>

(ex) 点推定の問題においては統計家は未知母数の値をあてようとする。<br>
この場合統計家の決定とは母数空間の元をなにか1つ選ぶことであるから決定空間は母数空間に一致し($D=\Theta$)、決定$d$は$ d \in \Theta$の元となる。<br>

(ex2) 検定の場合、統計家の取りうる決定は「棄却」および「受容」のいずれかである。<br>
従って決定空間 $D=${棄却, 需要}となる。棄却を1, 需要を0で表すと$D=\{0,1\}$の2点集合となる。


<font size=4>損失関数について</font><br>
統計的推測の問題を数学的な最適化問題に帰着させるための概念として損失関数がある。損失関数$L(\theta,d) \geq 0$は母数と決定の２つの変数の非負関数であり、<br>
損失関数の値$L=L(\theta,d)$は真のパラメータの値が$\theta$であるとき、統計家がとったある$d$という決定をとったときに統計家が被る損失の大きさを表すものである。<br>

(ex) 点推定では損失関数として二乗誤差$L(\theta,d)=(\theta-d)^2$が用いられることが多い。<br>
未知の$\theta$をあてようとするとき、推定値dがから離れれば離れるほど統計家の被る損失は大きくなると考えられるのでその点では損失関数の定義としてもっともらしい。<br>
ほかにも$L(\theta,d)=|\theta-d|$などが挙げられる。<br>
二乗誤差が用いられるのは数学的な取り扱いのしやすさにある。ただし、二乗誤差のもとで導かれる結論の多くはほかの定義の損失関数のもとでも成立する。<br>

(ex2)検定問題においては0-1損失という損失関数がわかりやすい。検定問題では母数空間$\Theta$が互いに排反な2つの部分集合の和として$\Theta = \Theta_0 \cup \Theta_1\;,\;\Theta_0\cap\Theta_1=\varnothing$と表されていると仮定する。<br>
決定空間は$D=${0,1}である。真のパラメータが$\Theta_0$に属するときの正しい決定は$d=0$であり、 真のパラメータが$\Theta_1$に属するときの正しい決定は$d=1$である。そのほかの場合には誤った決定をしたものと考える。<br>
正しい決定をした場合には損失は0であり、誤った決定をした場合には1の損失を被るとするのが0-1損失である。つまり、

---
(0-1損失)
$$L(\theta,\:0) = 
\begin{cases}
0, & if\; \theta \in \Theta_0 \\
1, & if\; \theta \in \Theta_1
\end{cases}$$

$$L(\theta,\:1) = 1-L(\theta,\:0)=
\begin{cases}
1, & if\; \theta \in \Theta_0 \\
0, & if\; \theta \in \Theta_1
\end{cases}$$

---

である。<br>
損失関数を導入することにより統計的推測の問題を統一的に定式化することができる。<br>
推定、検定を問わず統計的推測の目的は大雑把に言えば統計家にとって(平均的に)損失の少ない決定をおこなうことである。


<font size=4>決定関数について</font><br>
決定$d$について議論を深める。統計家は通常決定dを観測値Xに基づいて選ぶ。すなわち、dはXの関数$d=\delta(X)$であると言える。<br>
$\mathcal{X}$から$D$への関数$\delta$を決定関数あるいは決定方式という。<br>
決定関数とはすべての実現値$x$に対してとるべき決定$d=\delta(x)$を指定する関数である。<br>

(ex) 正規分布の母平均を標本平均で推定するとする。このときの決定関数は$\delta(x)=\bar{x}$である。<br>

以下、決定関数と決定方式の2つの用語が文脈上用いられることになるが、どちらも同じ意味で用いることとする。

<font size=4>リスク関数について</font><br>
決定関数は$X$の実数値関数であるから統計量である。従って決定関数の値$\delta(X)$は統計量の性質から確率変数であり、決定関数の値を用いて評価した損失関数$L(\theta,\delta(X))$の値も確率変数となる。<br>
つまり損失関数の値を小さくするといっても$\theta$が未知であり、確率的に変動する標本$X$のみに依存して決定$d$を選ぶ限り、損失関数の値も確率的に変動せざるを得ない。<br>
そこで平均的な損失を考えることにする。損失関数の期待値をリスク関数という。すなわちリスク関数とは次の式で定義される母数$\theta$と決定関数$\delta$の2つの変数を持つ関数である。

---
(リスク関数)
$$R(\theta, \delta)=E_\theta[L(\theta,\delta(X))]$$

---
ここで期待値は$P_\theta$に関するものであることに注意する。

(ex) 二乗誤差を用いた点推定の問題では$R(\theta,\delta)=E_\theta[(\theta-\delta(X))^2]$となる。この値を平均二乗誤差という。<br>

(ex2) 0-1損失を用いた検定問題においては$L(\theta,\delta(X))$は0か1かの2つの値しかとらないので$L$の期待値は$E[L]=0\times P_\theta(L=0)+1\times P_\theta(L=1)=P_\theta(L=1)$となる<br>
($P_\theta(L=0)$ : 母数の真の値がのとき、受容する確率)<br>
L=1とは誤った決定をした場合であるからリスク関数は<br>
$R(\theta,\delta)=
\begin{cases}
P_\theta(\delta(X)=1), & if \; \theta \in \Theta_0 \\
P_\theta(\delta(X)=0), & if \; \theta \in \Theta_1
\end{cases}$
 で与えられる

 統計的決定理論では決定関数の良し悪しをリスクの大小で比較する。すなわちリスクの小さい決定関数が望ましい決定関数であるということになる。<br>
 そしてリスクを最小にする決定関数を求めることが統計的決定理論の目的となる。またリスク関数の意味で望ましい決定方式を求める問題を統計的決定問題という。<br>


<font size=4>リスク関数を用いて議論することの問題点</font><br>
統計的推測の問題をリスク関数を用いて議論するということの問題点について触れる。<br>
決定方式をリスク関数で評価するということは、その決定方式を繰り返し用いたときの平均的な損失によって評価するということである。<br>
平均的な損失が小さければここの推測の問題においても損失の小さいことが期待できると考えるわけだが、平均的な良さは必ずしもここの問題での良さを保証するものではない。<br>
比喩を用いるとたとえば多くの人に効果のある薬は良い薬であると考えられるが、特定の個人には効果がなかったり、逆に害があるということもありうる。<br>

次により技術的な問題としてリスク関数は損失関数の選び方や母集団分布の想定に依存するという点が挙げられる。(リスク関数は仮定した母集団についての期待値であるから)<br>
例えば正規分布の仮定のもとで平均二乗誤差を基準として望ましい推定方式がほかの分布やほかの損失関数のもとでも望ましいという保証はない。<br>
この問題は剛健性(ロバストネス)の問題と呼ばれている。<br>

リスク関数を用いて決定方式を評価するとしても、それで問題が簡単に解決するわけではない。<br>
いま、2つの決定方式$\delta_1,\delta_2$についてリスク関数$R(\theta,\delta_i):i=1,2$を$\theta$の関数として考える。
